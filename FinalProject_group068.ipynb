### 1) Dropping Columns not needed in df_movies

```python
df_movies = df_movies.drop(labels=['title','date_published','duration', 'director', 'writer','production_company', 'actors', 'description', 'avg_vote', 'votes', 'worlwide_gross_income', 'metascore', 'reviews_from_users', 'reviews_from_critics'],
                          axis=1)
print("df_movies total (rows, columns): ", df_movies.shape)
df_movies.head()
```

### 2) Filtering out movies not from the USA and not in English from df_movies

```python
df_movies = df_movies[(df_movies['country']=='USA') & (df_movies['language']=='English')]
print("df_movies total (rows, columns): ", df_movies.shape)
df_movies.head()
```

### 3) Filtering out movies before 1990 from df_movies

```python
df_movies['year'] = df_movies['year'].replace('TV Movie 2019','2019')
df_movies['year'] = df_movies['year'].astype(int)
df_movies = df_movies[df_movies['year'] >= 1990]
print("df_movies total (rows, columns): ", df_movies.shape)
df_movies.head()
```

### 4) Filtering columns not needed in df_ratings

```python
df_ratings = df_ratings.loc[:,['imdb_title_id','weighted_average_vote']]
print("df_ratings total (rows, columns): ", df_ratings.shape)
df_ratings.head()
```

### 5) Merge ratings with valid titles into df_main

```python
df_main = pd.merge(df_movies,df_ratings, on='imdb_title_id')
print("df_main total (rows, columns): ", df_main.shape)
df_main.head()
```

### 6) Order the titles by year in df_main

```python
df_main = df_main.sort_values(by='year',ascending=True)
df_main = df_main.reset_index(drop=True)
print("df_main total (rows, columns): ", df_main.shape)
df_main.head()
```

### 7) Merge movies in df_budgets to df_main

```python
df_main = pd.merge(df_main, df_budget, how='left',on='original_title')

mask = df_main.isna()
df_main = df_main[~((mask.usa_gross_income == True)&(mask.Domestic_Gross == True))]
df_main

mask2 =df_main.isna()
df_main= df_main[~((mask2.budget == True)&(mask2.Production_Budget ==True))]
print("df_main total (rows, columns): ", df_main.shape)
df_main.head()
```

### 8) Drop columns not needed

```python
df_main['budget'] = df_main['budget'].fillna(df_main['Production_Budget'])
df_main['usa_gross_income'] = df_main['usa_gross_income'].fillna(df_main['Domestic_Gross'])
df_main = df_main.loc[:,['imdb_title_id','original_title','year','genre','weighted_average_vote','usa_gross_income']]
print("df_main total (rows, columns): ", df_main.shape)
df_main.head()
```

### 9) Standardizing income column

```python
df_main['usa_gross_income'] = df_main['usa_gross_income'].str.replace('$', '')
df_main['usa_gross_income'] = df_main['usa_gross_income'].str.replace(',', '')

df_main['usa_gross_income'] = df_main['usa_gross_income'].astype(int)
print("df_main total (rows, columns): ", df_main.shape)
df_main.head()
```

### 10) Delete duplicate rows and order by tidy data format

```python
df_main.drop_duplicates(subset='imdb_title_id', keep='first', inplace=True) #only keep eldest occurance
df_main = df_main.reset_index(drop=True) #reset the index of the table

df_main = df_main[['imdb_title_id', 'year', 'original_title', 'weighted_average_vote', 'usa_gross_income', 'genre']]
print("df_main total (rows, columns): ", df_main.shape)
df_main.head()
```

### 4a) Identify different genres

```python
def unique_genre(genres): 
    ans = [] 
    
    for one_row in genres: 
        each_genre = one_row.split(',')
        
        
        for x in each_genre: 
            x = x.strip()
            
            if x not in ans: 
                ans.append(x) 
            
    return ans 
```

### 4c) Count number of movies of genre in these two dataframes

```python
def num_movies(unique_genres, df):
    
    ans = []
    num_of_movies = 0

    for each_genre in unique_genres:
        
        for i in df.index:
            
            if each_genre in df.loc[i, 'genre']:
                num_of_movies += 1
      
        ans.append(num_of_movies)
        
        num_of_movies = 0
    
    return ans
```

## ANALYSIS 

From these datasets, we were able to match actors and actresses to their gender and the movies they are in (if they were in the top 10 most important contributors of the movie). Then, we were able to figure out which movies where the main cast was majority female and which ones were majority male. Within each genre, if the genre values surpassed the average calculated, we labeled the value as having changed/improved since before and after 2010. Depending on whether gender-diverse movies’ revenue and/or ratings increased over the years, we would consider each genre as being not successful, moderately successful, and successful in relation with gender diversity.
 
The following are steps we took to form the analysis:

### 1) Find gender diversity of each movie

```python
cast_count = pd.DataFrame(df_cast[['imdb_title_id', 'imdb_name_id']].groupby('imdb_title_id').count())
cast_count = cast_count.rename(columns={"imdb_name_id": "num_of_cast"})
cast_count.head()
```

```python
woman_count = pd.DataFrame(df_cast[['imdb_title_id','gender']].groupby('imdb_title_id').sum())
woman_count = woman_count.rename(columns={"gender": "num_of_woman"})
woman_count.head()
```

```python
movie_diversity = pd.merge(cast_count, woman_count, on='imdb_title_id')
movie_diversity['diversity'] = movie_diversity['num_of_woman'] / movie_diversity['num_of_cast']
#Creating a catergory that tells up if its diverse or not
for i in movie_diversity.index:
    if movie_diversity.loc[i,'diversity'] >= .5:
        movie_diversity.loc[i,'isDiverse'] = 1
    else:
        movie_diversity.loc[i,'isDiverse'] = 0
        
movie_diversity.head()
```

### 5) Find average diversity of each genre of movie in pre2010 and post2010

```python
def avg_diversity(unique_genres, df):
    
    ans = []
    sum_diversity = 0
    num_of_movies = 0

    for each_genre in unique_genres:
        
        for i in df.index:
            
            if each_genre in df.loc[i, 'genre']:
                sum_diversity += df_main.loc[i, 'diversity']
                num_of_movies += 1

        if num_of_movies == 0:
            ans.append(0)
        else:
            ans.append(sum_diversity/num_of_movies)
        

        sum_diversity = 0
        num_of_movies = 0
    
    return ans
```

```python
analysispost2010['avg_diversity'] = avg_diversity(unique_genres, post2010)
analysispost2010
```

 ### 6) Find average rating of each genre of movie in pre2010 and post2010

```python
def avg_rating(unique_genres, df):
    
    ans = []
    sum_ratings = 0
    num_of_movies = 0
    

    for each_genre in unique_genres:
        
        for i in df.index:
            
            if each_genre in df.loc[i, 'genre']:
                sum_ratings += df_main.loc[i, 'weighted_average_vote']
                num_of_movies += 1

        if num_of_movies == 0:
            ans.append(0)
        else:
            ans.append(sum_ratings/num_of_movies)
            
        sum_ratings = 0
        num_of_movies = 0
    
    return ans
```

```python
analysispre2010['Pavg_rating'] = avg_rating(unique_genres, pre2010)
analysispre2010
```

```python
analysispost2010['avg_rating'] = avg_rating(unique_genres, post2010)
analysispost2010
```

### 7) Find woman-diverse movie's average ratings pre2010 and post2010

```python
def woman_diverse_ratings(unique_genres, df):
    
    ans = []
    sum_average = 0
    num_of_movies = 0

    for each_genre in unique_genres:
        
        for i in df.index:
            
            if each_genre in df.loc[i, 'genre']:
                
                if df.loc[i, 'diversity'] >= 0.5:
                    sum_average += df_main.loc[i, 'weighted_average_vote']
                    num_of_movies += 1

        if num_of_movies == 0:
            ans.append(0)
        else:
            ans.append(sum_average/num_of_movies)
        
        sum_average = 0
        num_of_movies = 0
    

    return ans
```

```python
analysispre2010['Pwoman_avg_rating'] = woman_diverse_ratings(unique_genres, pre2010)
analysispre2010
```

```python
analysispost2010['woman_avg_rating'] = woman_diverse_ratings(unique_genres, post2010)
analysispost2010
```

### 8) Find avg gross income per genre pre2010 and post2010

```python
def avg_income(unique_genres, df):
    
    ans = []
    sum_income = 0
    num_of_movies = 0
    

    for each_genre in unique_genres:
        
        for i in df.index:
            
            if each_genre in df.loc[i, 'genre']:
                sum_income += df_main.loc[i, 'usa_gross_income']
                num_of_movies += 1

        if num_of_movies == 0:
            ans.append(0)
        else:
            ans.append(sum_income/num_of_movies)
            
        sum_income = 0
        num_of_movies = 0
    
    return ans
```

## ETHICS & PRIVACY
Our goal is to collect data from trustworthy, public sites without any inherent biases. Here are ways we ensured the highest quality of data ethics and privacy in our project for the different types of data we worked with:

This information is safe to use for the purpose of our project since we do not have any intentions of defamation for any observations used in our datasets. To take care of any outward biases, we need to fact-check the source we get this information from to make sure it is accurate and not misleading or faulty.  This information does not have any other potential issues with data ethics and privacy as long as we follow the guidelines explained above. 
 
 
#### Movie Ratings and IMBd Data
We gathered our movie ratings from the service IMDb for the purpose of our project since we are analyzing it for personal and non commercial use only. It is important that we consult the service’s terms and conditions to double check that we comply with their permissions. The greatest privacy concern regarding this data is that we do not invade any individual viewer information. IMDb publishes the weighted vote averages rather than the raw data averages. Additionally, they accept and consider all votes received by the users, but not all votes have the same impact on the final rating. We are aware that there is no foolproof way for IMBd to verify that a voter has viewed a movie before actually rating it. That being said, some “controversial” films may contain a skewed score from positive or negative “ballot-stuffing” (although IMDb does apply several countermeasures to unskew ratings). The ratings from IMDb are all user submitted and therefore do not come from other external sources. A film may have lesser amounts of votes or a lower score due to several factors. These factors could potentially be that not all people are willing to vote or register to rate a movie they do not necessarily feel passionate about or that a film’s demographic does not use IMDb.
