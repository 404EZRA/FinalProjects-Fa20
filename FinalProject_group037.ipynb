## Background and Prior Work
Gacha games are games that use the gacha system which originated from “gachapon” systems, or toy vending machine mechanics, where one puts in money in order to get a chance at receiving a random item. This system is an incredibly lucrative business model in digital games because players are often blinded by their optimism to gain special items on each roll they perform. There are generally two types of gacha games: one where the game has the gacha system as its core where rolled items give you better stats or characters, and the other where games have the gacha system only for commercial purposes like selling aesthetic items that don't change character stats. Our research question focuses on the former type, and the specific game we will research on is Fate Grand Order, a successful and popular mobile game that is representative of its type, namely, anime-style mobile gacha game. Here, the rarest 5-star characters are called SSR. All SSRs are either non-limited or limited, where the former could be obtained all-year-round and the latter is only available in timely and limited summoning banners. Characters are rolled randomly using in-game currencies converted from real money rather than directly purchased with a fixed price. Therefore, players have a chance to get a rare item with much less cost than it should be, but there’s also a chance that it’d cost a lot more. Like most mobile gacha games, Fate/Grand Order is a free-to-play game that exploits this strategy to make players spend real-world currency. We’re interested in how specific features of this gacha game could affect the players’ willingness to pay.

**Measurements:**
- Exclusivity:
  The exclusivity of a character is measured by how many times it’s been rate-up since fgo started to operate until the time of the survey. Rate-up means there is an additional summoning banner opened featuring this character. Note that exclusivity only applies to limited SSR characters. Only these characters are made exclusive this way because players can’t obtain them when the featured banners aren’t open. Non-limited SSRs are available all year through the normal summoning banner, thus although their number of times being rated up is also counted, the concept of exclusivity doesn’t apply to them.


- Power-level:
  The power level of a character is measured by the tier lists made by authoritative game medias like Appmedia and Gamepress at the time when this survey is conducted, ranking characters in different tiers based on their power-levels. This is the most objective way we came up with considering the partially-subjective nature of this feature. The ranking from appmedia represents the power tiers in JP server at the time when this survey is conducted, and the ranking from gamepress represents the rankings in NA server at the time. The JP server is two years ahead of the NA server in terms of game progress, so the power-level in JP server can be used to assess whether NA players consider future power-levels when deciding which character to invest in and pull for. We then assigned weights for each character based on this data. Specific scales are mentioned in data cleaning.


- Likeability:
  Likeability refers to how likable a character is based on its character setting, personality, appearance and any other features unrelated to game stats. Because the source of a character’s popularity could be roughly divided into two parts, namely game-stats and likeability, we originally wanted to examine to what extent a character’s likeability alone would increase players’ willingness to pay. At first it was represented by the ratio obtained from dividing the possession number of each character over its power-level, but we then thought this approach may be flawed since it doesn’t accurately reflect likeability when the possession and power share the same trend, for example a character with possession and power both at 4 have the same ratio as a character with both features at 1, but they obviously don’t share the same state. Therefore, our final approach is normalizing a character’s possession number to the same scale of power first and then subtract power from it. This way the trade off between power-level and likeability is better reflected.


- Players’ willingness to pay:
  Because limited SSRs could be only acquired through timely and limited summoning banners, we use the possession rate of each character and the total possession of SSR in each spending group to represent the players’ willingness to pay.

Next we try ridge and lasso regression to see if applying regularization changes this result. Linear regression model in sklearn.linear_model doesn’t have the feature to generate a summary table, so we'll choose OLS instead and apply regularizations there.


```python
#lasso on overall model
names = ['F2P_percent_total','Paid_Gacha_only_percent_total','Occasional_Spender_percent_total','Moderate_Spender_percent_total','Low_High_Spender_percent_total','High_Spender_percent_total','High_High_Spender_Whaler_percent_total']
view = []
for name in names:
    X = np.array(data_limited[['Rate Up (Exclusivity)','Power JP','Power NA']])
    Y = np.array(data_limited[name])
    scaler = StandardScaler().fit(X)
    X = scaler.transform(X)
    X = sm.add_constant(X) 
    
    model = linear_model.OLS(Y,X)
    results_fu = model.fit()
    results_fr = model.fit_regularized(start_params=results_fu.params)
    final = sm.regression.linear_model.OLSResults(model, 
                                              results_fr.params, 
                                              model.normalized_cov_params)

    view.append([final,name])

# print results    
for i in range(7):
    print(view[i][0].summary())#.get_params)
```

When we're doing Lasso regression instead, all three features are not siginificantly correlated with possession rate. Power-NA is still the most relatable feature indeed, but the p values are still slightly bigger than 0.01.

# Conclusion & Discussion

- **Data and question:** Overall, we examined character-based features like exclusivity and power-level of SSR characters in Fate/ Grand Order and player-based features like total time played and gender to determine what factor increases players’ willingness to pay the most among different spending groups. We used survey data from reddit that incorporated over three thousand observations collected from active players. 


- **Analysis:** We first applied correlation matrix to learn correlations between two single features. Then we applied linear regression models with various regularizations to learn what features are statistically significantly correlated with SSR possession through p values and how they are correlated in coefficients.    


- **Results and conclusion:** 
  Based on the results, we reject our hypothesis for character-based features. At least for this sample, only power-level in the NA server shows a significant positive correlation with players’ willingness to pay, meaning that American players still make character pulling decisions largely based on game-stats even if this is just a mobile-gacha game whose emphasis isn’t at gameplays, and this is true for all spending groups. The most commonly-possessed character list also confirms this conclusion, since its top 3 are all among the highest-ranked characters. Power-level in the JP server doesn’t show such significant correlation, meaning that American players only care about current power-levels and rarely care about the future power-level of a certain character when making decisions on which character to pull. There is no sign of investing in the future here. And to our surprise, exclusivity has no statistically significant effect on willingness to pay, which is a bit counter-intuitive because we would generally think that the more exclusive an item is, the more willing players are to pay and roll for it. It turns out the US players are quite reasonable in this by not assigning more value to a rarer item. The final ranking of importance here is power NA, power JP and exclusivity, from the most influential to the least influential. Because Fate/Grand Order is one of the most successful anime-style mobile gacha games and a representation of its type, we can generalize that even for these games that put character likeabilities rather than gameplay as their greatest selling point, players still largely treat them as stats games and make pulling decisions largely based on character power-levels. 

  We also reject our hypothesis on likeability since there is no significant correlation shown between likeability and rate-up times, suggesting that the developer doesn't purposely arrange rate-ups based on how likeable a character is, probably because they don’t see enough potential profit here.  

  Lastly, we partially reject our hypothesis on player-based features. Total time played proved to be the most influential factor indeed with a positive correlation, but gender turns out to be an uncorrelated feature, meaning that players of different genders don’t show any significant divergence in willingness to pay and fluctuation in the amount of male players don’t affect the willingness to pay either within the entire player base. Other factors are all significantly correlated with willingness to pay but not like the way we hypothesized. Whether saved for a specific character shows a positive correlation, meaning that this kind of player is more emotionally invested in the game and would be more likely to pay for characters of interest. Yet to our surprise, faith in the game developer and whether rerolled demonstrated negative correlation, suggesting that players who acknowledge the effort and caring of the game developer and players who have rerolled for a better starter account tend to spend less money in the game. The former may be because these players believe the developer would continue to care about the low-spending group so that they could still play fine without paying much, and the latter may be because the more strong limited characters a rerolled account has, the less need exists for the player to pay for new characters.
  
  In summary, for players in NA server, the most influential features in Fate Grand Order that increase players' willingness to pay are a character's current power-level and the total time played.
