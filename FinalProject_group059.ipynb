To begin, we first start with updating the `state_marj_laws2016_df` to a more current `state_marj_laws2020_df` determine which states have legalized commercial cannabis production and retail sales. Alongside this, let us also develop a function to make our printed dataframes truly pleasing to the eye.

```python
# DEF: Output dataframe as "pretty styled" image.
def render_mpl_table(data, col_width=3.0, row_height=0.625, font_size=14,
                     header_color='#40466e', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')

    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)

    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in  six.iteritems(mpl_table._cells):
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ])
    return ax

# Replace Nan variables with false
state_marj_laws2020_df = state_marj_laws2016_df.replace({np.nan: False, 'Yes': True})

# Updated list of states that are now recreational *Scraped from Wikipedia (https://en.wikipedia.org/wiki/Legality_of_cannabis_by_U.S._jurisdiction)
update_recreational_states = ['Arizona','Illinois','Maine','Massachusetts','Michigan','Montana','New Jersey','South Dakota','Vermont','Washington']

# Rename all state_marj_laws2020_df column names to something more useful/concise and convert to lowercase
state_marj_laws2020_df = state_marj_laws2020_df.rename({'Medical marijuana legalized':'medicinal'}, axis='columns')
state_marj_laws2020_df = state_marj_laws2020_df.rename({'Marijuana legalized for recreational use':'recreational'}, axis='columns')
state_marj_laws2020_df = state_marj_laws2020_df.rename({'No laws legalizing marijuana':'illegal'}, axis='columns')
state_marj_laws2020_df = state_marj_laws2020_df.rename(columns=str.lower)

# Iterate through state_marj_laws2020_df and update laws
for ind in update_recreational_states:
    state_marj_laws2020_df.loc[state_marj_laws2020_df['state'] == ind, ['medicinal','illegal']] = False
    state_marj_laws2020_df.loc[state_marj_laws2020_df['state'] == ind, ['recreational']] = True

# Drop all observations (states) from dataframe tha have not legalized the recreational use of marijuana. Sort and reindex the dataframe
pd.set_option("display.max_rows", None, "display.max_columns", None)
state_marj_laws2020_df = state_marj_laws2020_df.sort_values(by=['recreational','state'],ascending=True)
state_marj_laws2020_df = state_marj_laws2020_df.reset_index(drop=True)

state_marj_laws2020_df = state_marj_laws2020_df[state_marj_laws2020_df.recreational != False]
state_marj_laws2020_df = state_marj_laws2020_df.reset_index(drop=True)

# Display dataframe to see which states have recreational marijuanas
render_mpl_table(state_marj_laws2020_df,header_columns=0,col_width=2, row_height=0.525, font_size=13)
```

```python
# DEF: Convert to lowercase all column names of input df
def lowercase_column_names(df): 
    df.columns = map(str.lower, df.columns)
    return df

# DEF: Drop rows with repeating years and rescale; developed in mind for agesex_##_state_df dataframes
def drop_repeat_year_row(df, alert_status=None):
    # Drop all rows that pertain to the year "2" e.g., the date 4/1/2020
    df = df[df.year != 1]
    # Drop all rows that pertain to the year "2" e.g., the date 4/1/2020
    df = df[df.year != 2]
    # Rescale by subtracting all year values by 2
    df = df.apply(lambda x: np.subtract(x,2) if x.name in ['year'] else x)
    # Reset index of modfied dataframe
    df = df.reset_index(drop=True)
    return df

# DEF: Vectorize rows of same county; developed in mind for agesex_##_state_df dataframes
def condense_time_to_row(df, alert_status=None):
    numRows = len(df.index)
    
    # Create beginning block of result dataframe
    temp_df_0 = df.iloc[[0],6:]
    temp_df_0 = temp_df_0[0:9].add_suffix("_0")
    temp_df_1 = df.iloc[[1], 6:]
    temp_df_1 = temp_df_1[0:9].add_suffix("_1")
    temp_df_2 = df.iloc[[2], 6:]
    temp_df_2 = temp_df_2[0:9].add_suffix("_2")
    temp_df_3 = df.iloc[[3], 6:]
    temp_df_3 = temp_df_3[0:9].add_suffix("_3")
    temp_df_4 = df.iloc[[4], 6:]
    temp_df_4 = temp_df_4[0:9].add_suffix("_4")
    temp_df_5 = df.iloc[[5], 6:]
    temp_df_5 = temp_df_5[0:9].add_suffix("_5")
    temp_df_6 = df.iloc[[6], 6:]
    temp_df_6 = temp_df_6[0:9].add_suffix("_6")
    temp_df_7 = df.iloc[[7], 6:]
    temp_df_7 = temp_df_7[0:9].add_suffix("_7")
    temp_df_8 = df.iloc[[8], 6:]
    temp_df_8 = temp_df_8[0:9].add_suffix("_8")
    temp_df_9 = df.iloc[[9], 6:]
    temp_df_9 = temp_df_9[0:9].add_suffix("_9")
    temp_df = pd.concat([temp_df_0,temp_df_1,temp_df_2,temp_df_3,temp_df_4,temp_df_5,temp_df_6,temp_df_7,temp_df_8,temp_df_9], axis=0, sort=False)
    temp_df = temp_df.stack().to_frame().T
    
    # Build list of right column header strings that are "correct"  
    newColHeadersRight = [lis[1] for lis in temp_df.columns] 
    
    # Rename column headers to simply counting increments
    temp_df.columns = [np.arange(0,temp_df.shape[1])]
    
    for i in range(9,numRows): 
        if df.loc[i,"year"] == 1 :
            jEnd = i + 10
            temp_dfSlice = df.iloc[i:jEnd,6:]
            temp_dfSlice = temp_dfSlice.stack().to_frame().T
            temp_dfSlice.columns = [np.arange(0,temp_dfSlice.shape[1])]
            temp_df = pd.concat([temp_df, temp_dfSlice], axis=0, sort=False)
    
    df = df.iloc[:,0:6]
    df = df[df.year == 1]
    df = df.iloc[:,0:5]
        
    # Build list of left column header strings that are "correct"
    newColHeadersLeft = list(df.columns) 
    
    df.columns = [np.arange(0,df.shape[1])]
    temp_df.columns = [np.arange(0,temp_df.shape[1])]
    df = df.reset_index(drop=True)
    temp_df = temp_df.reset_index(drop=True) 
    
    result = pd.concat([df, temp_df], axis=1)
    result.columns = [np.arange(0,result.shape[1])]
    
    # Rename column headers to correct using list of "correct" stirngs
    oldColHeaders = [lis[0] for lis in result.columns] 
    
    newColHeaders = newColHeadersLeft + newColHeadersRight
    
    # Rename column headers using "correct" strings
    newDict = dict(zip(oldColHeaders, newColHeaders))
    result = result.rename(columns=newDict)
    
    return result
                

# Convert all column headers to lower case
agesex_02_alaska_df = lowercase_column_names(agesex_02_alaska_df)
agesex_04_arizona_df = lowercase_column_names(agesex_04_arizona_df)
agesex_06_california_df = lowercase_column_names(agesex_06_california_df)
agesex_08_colorado_df = lowercase_column_names(agesex_08_colorado_df)
agesex_17_illinois_df = lowercase_column_names(agesex_17_illinois_df)
agesex_23_maine_df = lowercase_column_names(agesex_23_maine_df)
agesex_25_massachusetts_df = lowercase_column_names(agesex_25_massachusetts_df)
agesex_26_michigan_df = lowercase_column_names(agesex_26_michigan_df)
agesex_30_montana_df = lowercase_column_names(agesex_30_montana_df)
agesex_32_nevada_df = lowercase_column_names(agesex_32_nevada_df)
agesex_34_new_jersey_df = lowercase_column_names(agesex_34_new_jersey_df)
agesex_41_oregon_df = lowercase_column_names(agesex_41_oregon_df)
agesex_46_south_dakota_df = lowercase_column_names(agesex_46_south_dakota_df)
agesex_50_vermont_df = lowercase_column_names(agesex_50_vermont_df)
agesex_53_washington_df = lowercase_column_names(agesex_53_washington_df)

# Drop all unnecessary observations of repeated years
agesex_02_alaska_df = drop_repeat_year_row(agesex_02_alaska_df)
agesex_04_arizona_df = drop_repeat_year_row(agesex_04_arizona_df)
agesex_06_california_df = drop_repeat_year_row(agesex_06_california_df)
agesex_08_colorado_df = drop_repeat_year_row(agesex_08_colorado_df)
agesex_17_illinois_df = drop_repeat_year_row(agesex_17_illinois_df)
agesex_23_maine_df = drop_repeat_year_row(agesex_23_maine_df)
agesex_25_massachusetts_df = drop_repeat_year_row(agesex_25_massachusetts_df)
agesex_26_michigan_df = drop_repeat_year_row(agesex_26_michigan_df)
agesex_30_montana_df = drop_repeat_year_row(agesex_30_montana_df)
agesex_32_nevada_df = drop_repeat_year_row(agesex_32_nevada_df)
agesex_34_new_jersey_df = drop_repeat_year_row(agesex_34_new_jersey_df)
agesex_41_oregon_df = drop_repeat_year_row(agesex_41_oregon_df)
agesex_46_south_dakota_df = drop_repeat_year_row(agesex_46_south_dakota_df)
agesex_50_vermont_df = drop_repeat_year_row(agesex_50_vermont_df)
agesex_53_washington_df = drop_repeat_year_row(agesex_53_washington_df)

# Vectorize population dataframes with respect to time
agesex_02_alaska_df = condense_time_to_row(agesex_02_alaska_df) 
agesex_04_arizona_df = condense_time_to_row(agesex_04_arizona_df)
agesex_06_california_df = condense_time_to_row(agesex_06_california_df)
agesex_08_colorado_df = condense_time_to_row(agesex_08_colorado_df)
agesex_17_illinois_df = condense_time_to_row(agesex_17_illinois_df)
agesex_23_maine_df = condense_time_to_row(agesex_23_maine_df)
agesex_25_massachusetts_df = condense_time_to_row(agesex_25_massachusetts_df)
agesex_26_michigan_df = condense_time_to_row(agesex_26_michigan_df)
agesex_30_montana_df = condense_time_to_row(agesex_30_montana_df)
agesex_32_nevada_df = condense_time_to_row(agesex_32_nevada_df)
agesex_34_new_jersey_df = condense_time_to_row(agesex_34_new_jersey_df)
agesex_41_oregon_df = condense_time_to_row(agesex_41_oregon_df)
agesex_46_south_dakota_df = condense_time_to_row(agesex_46_south_dakota_df)
agesex_50_vermont_df = condense_time_to_row(agesex_50_vermont_df)
agesex_53_washington_df = condense_time_to_row(agesex_53_washington_df)

# Combine the various state population dataframes into one comprehensive dataframe
agesex_usa_df = pd.concat([agesex_02_alaska_df  ,agesex_04_arizona_df ,agesex_06_california_df ,agesex_08_colorado_df ,agesex_17_illinois_df ,agesex_23_maine_df ,agesex_25_massachusetts_df ,agesex_26_michigan_df ,agesex_30_montana_df ,agesex_32_nevada_df ,agesex_34_new_jersey_df ,agesex_41_oregon_df ,agesex_46_south_dakota_df ,agesex_50_vermont_df ,agesex_53_washington_df ], axis=0, sort=False)

# Reset index of agesex_usa_df dataframe
agesex_usa_df = agesex_usa_df.reset_index(drop=True)

# Preview dataframe
pd.set_option('display.max_rows', 5)
display(agesex_usa_df)
```

Because there are a lot of factors to consider when answering our question, we have evaluated our project based on several efficient measurements.
